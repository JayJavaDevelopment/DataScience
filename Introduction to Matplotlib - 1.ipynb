```python
over_50 = heart_disease[heart_disease['age']>50]
over_50
```

```python
type(over_50)
```

```python
over_50.plot(x='age',y='chol',kind='scatter',figsize=(10,6),c='target')
```

```python
fig, ax = plt.subplots(figsize=(10,6) )
over_50.plot(x='age',y='chol',kind='scatter',c='target',ax=ax)
ax.set_xlim([45,100])
##Basic plot
```

```python
over_50
```

```python
## Defining the graph size
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(over_50["age"],
                        over_50["chol"],
                         c=over_50["target"])

ax.set(title="Heart Disease and Cholesterol Levels",
          xlabel="Age",
          ylabel="Cholestral")
ax.legend(*scatter.legend_elements(), title="Target");

##Adding a mean line
ax.axhline(over_50["chol"].mean(),linestyle="--",linewidth=1,c='red')
```

```python
## Defining the graph size

fig, (ax0,ax1) = plt.subplots(figsize=(10,6),nrows=2,ncols=1)

##Plotting first graph
# Plot the data
scatter = ax0.scatter(over_50["age"],
                        over_50["chol"],
                         c=over_50["target"])

ax0.set(title="Heart Disease and Cholesterol Levels",
          xlabel="Age",
          ylabel="Cholestral")
ax0.legend(*scatter.legend_elements(), title="Target");

##Adding a mean line
ax0.axhline(over_50["chol"].mean(),linestyle="--",linewidth=1,c='red')

##Plotting second graph

# Plot the data
scatter = ax1.scatter(over_50["age"],
                        over_50["thalach"],
                         c=over_50["target"])

ax1.set(title="Heart Disease and Cholesterol Levels",
          xlabel="Age",
          ylabel="thalach")
ax1.legend(*scatter.legend_elements(), title="Target");

##Adding a mean line
ax1.axhline(over_50["thalach"].mean(),linestyle="--",linewidth=1,c='blue')
```
