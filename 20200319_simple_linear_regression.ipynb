```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Jayaprakash\DataScience\Projects\Test Data\Simple_Linear_Salary_Data.csv")

X = df.iloc[:,:-1].values
y = df.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)

model = LinearRegression()
model.fit(X_train ,y_train)

y_pred = model.predict(X_test)
y_pred
## Out Model predicted for test data
```

```python
"""
Graphically displaying how is out model predicting for the training data itself (Not for test data)
"""

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Vs Title")
plt.scatter(X_train,y_train, c= 'red')
plt.plot(X_train,model.predict(X_train) );
```

```python
"""
Graphically displaying how is out model predicting for the test data.
"""
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Vs Title")
plt.scatter(X_test,y_test, c= 'red')
plt.plot(X_test,model.predict(X_test) );
```
