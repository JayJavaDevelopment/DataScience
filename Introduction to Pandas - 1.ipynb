# Manipulating Data

```python
car_sales
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing["Odometer"].mean()
```

```python
##Without inplace 
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
car_sales_missing
##Change in data displayed
```

```python
##With inplace 
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace = True)
car_sales_missing
```

```python
car_sales_missing.dropna(inplace = False)
car_sales_missing
```

```python
car_sales_missing.dropna(inplace = True)
car_sales_missing
```

```python
car_sales_missing_dropped = car_sales_missing
```

```python
car_sales_missing_dropped.to_csv("car_sales_missing_dropped" , index = False)
```

```python
seats = pd.Series([5,6,7,8,9])
```

```python
car_sales["Seats"].fillna(3,inplace = True)
car_sales
```

```python
## i'm using list and list should match the exact no of items it not error will be thrown
fuel_eco = [2.4,5.6,6,8,3,9,2,8,6.7,7.2]
```

```python
car_sales["Milake"] = fuel_eco
car_sales
```

```python
car_sales["To Drop"] = car_sales["Odometer (KM)"]/10 * car_sales["Milake"]/5
car_sales["Fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Milake"]
car_sales
```

```python
car_sales.drop("To Drop" , inplace = True , axis = 1)
car_sales
```
